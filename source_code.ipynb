```python
import matplotlib.pyplot as plt

# Data
months = ['January', 'February', 'March', 'April', 'May']
sales = [12000, 15000, 18000, 21000, 19000]
tickets = [100, 120, 140, 160, 130]

# Line chart for sales
plt.figure(figsize=(10, 5))
plt.plot(months, sales, marker='o', color='blue', label='Sales')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.legend()
plt.show()

# Bar chart for support tickets
plt.figure(figsize=(10, 5))
plt.bar(months, tickets, color='orange', label='Support Tickets')
plt.title('Support Tickets per Month')
plt.xlabel('Month')
plt.ylabel('Tickets')
plt.grid(True)
plt.legend()
plt.show()
```
